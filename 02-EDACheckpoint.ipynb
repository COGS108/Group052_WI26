# Research Question

Does playing back-to-back NBA games affect team performance? Using data from 2000 to the present, how do key performance metrics—such as points scored, field goal percentage, turnovers, and win/loss outcome—differ between games played on consecutive days and games with at least one day of rest?

## Background and Prior Work

Basketball is a high intensity, physically demanding sport played professionally in the National Basketball Association (NBA), where teams compete in an 82-game regular season typically running from October to April. During this period, each NBA team plays almost daily, often traveling long distances between cities. Over a season, this results in periods of “schedule congestion”, including back-to-back games, where a team plays on consecutive days and games with varying amounts of rest and recovery in between. Performance in these games is usually measured using team-level statistics such as points scored, field goal percentage (FG%), turnovers and the win/loss outcome, as well as more nuanced metrics like net efficiency, effective field goal percentage (an adjusted shooting measure accounting for three-point shots) and pace of play. These metrics provide insight into both offensive and defensive performance in comparison across games and conditions.<a name="cite_ref-1"></a>[<sup>1</sup>](#cite_note-1), <a name="cite_ref-2"></a>[<sup>2</sup>](#cite_note-2)

To determine which statistics most strongly relate to team success, prior empirical research has analyzed thousands of NBA games to identify performance indicators associated with winning outcomes. One large-scale research study examining nearly 4,000 NBA regular and postseason games found that field goal percentage and overall shooting efficiency were among the most influential variables distinguishing winning and losing teams.<a name="cite_ref-3"></a>[<sup>3</sup>](#cite_note-3) Defensive rebounding and shooting efficiency accounted for a substantial portion of the explained variance in game outcomes. The study also observed that teams tend to adopt more conservative styles of play under higher pressure conditions, resulting in fewer field goal attempts, assists, and turnovers. These findings establish that performance metrics such as FG%, scoring efficiency, and turnovers are not arbitrary statistics but empirically supported indicators of team success. At the same time, the authors acknowledge that these relationships are also statistical associations and do not imply that any single metric alone determines game outcomes, highlighting the multifactorial nature of basketball performance.

In addition to peer-reviewed research, prior student-led data science projects have explored NBA performance statistics using structure/approaches that are closely related to the project proposed here. ### Data Wrangling/Cleaning

Key goals:
- Remove non-regular / malformed rows that could bias team performance summaries

```python
## YOUR CODE TO LOAD/CLEAN/TIDY/WRANGLE THE DATA GOES HERE
import pandas as pd
import numpy as np

# Load and assign the raw data frame to tstats_df 
tstats_df = pd.read_csv('data/00-raw/TeamStatistics.csv')

# Removes entries with NaN in "teamCity" column, effectively removing any
# non-regular NBA games, like All-Star games.
tstats_df = tstats_df.dropna(subset=["teamCity"])

# Back-to-back classification depends on day-to-day differences between games, so convert timestamp to datetime (YYYY-MM-DD)
# To help, we eliminate the time component of the "gameDateTimeEst" column, which is not needed for our analysis
tstats_df["gameDateTimeEst"] = pd.to_datetime(tstats_df["gameDateTimeEst"]).dt.date

# Our research question focuses on modern NBA play (2000+) so want to
# begin looking at data from 2000's onwards, so we filter out any games before 2000.
tstats_df = tstats_df[tstats_df["gameDateTimeEst"].apply(lambda x: x.year) >= 2000]

# We focus on key metrics we plan to compare under rest vs back-to-back conditions:
# points scored, opponent points, FG%, turnovers, win/loss, and home/away
columns_to_keep = [
    'gameId',
    'gameDateTimeEst',
    'teamId',
    'teamName',
    'teamCity',
    'opponentTeamId',
    'home',
    'win',
    'teamScore',
    'opponentScore',
    'fieldGoalsPercentage',
    'turnovers'
]

# make tstats_df equal to a susbset of the original dataframe, keeping only the comlumns in columns_to_keep
tstats_df = tstats_df[columns_to_keep]

# Tail shows the last 10 rows of the dataframe, which should be the oldest games in the dataset.
tstats_df.tail(10)
```

### Data Integrity
To ensure the dataset structure accurately represents one team’s performance per game and is valid for back-to-back calculations, we verified that each gameId appears exactly twice (once per team) in the dataset. In the NBA data, each game consists of exactly two teams, so each game should generate exactly two rows, one for each team’s box score. The results of the code below confirms that all 35,656 games appear exactly twice, indicating that no games are missing a team entry and no games incorrectly include more than two teams. We also confirmed that there are no duplicate (gameId, teamId) rows to ensure that no team's performance is recorded more than once for the same game. Duplicate rows would inflate averages, distort win rates, and lead to incorrect rest-day calculations. Our check confirms zero duplicate team-game observations.

```python
# Each row should represent ONE team's stats in ONE game.
# That means each gameId should appear exactly twice (two teams per game).
game_rows = tstats_df.groupby("gameId").size()
print("Rows per gameId (counts of counts):")
print(game_rows.value_counts().sort_index())

# There should be no duplicate (gameId, teamId) pairs.
dup_game_team = tstats_df.duplicated(subset=["gameId", "teamId"]).sum()
print("\nDuplicate (gameId, teamId) rows:", dup_game_team)
```

Additionally, we examined whether any teamId maps to multiple teamName values and observed that a few teamIds map to multiple historical teamNames due to franchise renaming. Since all grouping, sorting, and rest calculations are performed using the stable teamId rather than teamName, these historical naming changes do not affect our back-to-back computations and analysis.

Because back-to-back identification depends on chronological ordering of games within each team, any missing or duplicated team-game observation would directly bias the rest-day variable. These structural checks therefore serve as a prerequisite for valid temporal analysis.

```python
# Team names can change over time (franchise renames), but teamId is the stable identifier.
team_name_counts = (
    tstats_df.groupby("teamId")["teamName"]
    .nunique()
    .sort_values(ascending=False)
)

print("\nTop teamIds with multiple teamNames (often due to franchise renaming):")
print(team_name_counts.head(10))

# List examples of teamIds that have multiple names + what those names are
multi_name_teamids = team_name_counts[team_name_counts > 1].index

# If there are teamIds that share multiple names, this code runs
if len(multi_name_teamids) > 0:
    print("\nExamples of historical renames (teamId -> teamNames observed):")
    examples = (
        # extracts only the rows teamId that is known to have multiple team names
        tstats_df[tstats_df["teamId"].isin(multi_name_teamids)]
        # groups the rows by teamId and keep only teamName column
        .groupby("teamId")["teamName"]
        # removes duplicate names within each group 
        .unique()
    )
    # print a few examples
    for team_id in list(examples.index)[:10]:
        print(team_id, "->", list(examples.loc[team_id]))
```

#### Identifying Back-to-Back Games

To address our research question, we first construct a variable that identifies whether a game
was played on the second night of a back-to-back.

We define a back-to-back game as one where a team plays on consecutive calendar days.
To compute this, we:
1. Sort each team's games chronologically.
2. Compute the difference in days between consecutive games.
3. Label games with exactly 1 day of rest as B2B = 1.
4. All other games are labeled B2B = 0.

This variable allows us to compare performance metrics under different rest conditions.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load cleaned dataset
df = pd.read_csv("data/02-processed/TeamStatistics_Cleaned.csv")

# Preview data
df.head(10)
```

```python
df.info()
```

```python
df.groupby('teamId').count() # 4 teams have 3 or less games in the dataset
team_counts = df.groupby('teamId').size()
team_counts.sort_values().head(10)
```

```python
# remove 4 teams so it cannot skew our analysis and we cannot compute back to back with only 1 game
team_counts = df["teamId"].value_counts()
valid_teams = team_counts[team_counts > 3].index

# checking correct amount of teams
len(team_counts)
len(valid_teams)
valid_teams
```

```python
df = df[df["teamId"].isin(valid_teams)]

# convert to 'gameDataeTimeEst' to datetime64 (object type before)
df["gameDateTimeEst"] = pd.to_datetime(df["gameDateTimeEst"])
df.dtypes
```

```python
# reorders the dataframe by the teamIds, then the game times
df = df.sort_values(["teamId", "gameDateTimeEst"]).reset_index(drop = True)
df.head(10)
```

```python
# calculating back to back games
# computes how many days since the last game, 1 means back to back, >1 not back to back
df["rest_days"] = df.groupby("teamId")["gameDateTimeEst"].diff().dt.days 
# 1 means back to back, 0 means not back to back
df["B2B"] = (df["rest_days"] == 1).astype(int) 
df["B2B"].value_counts()
```

```python
# comparing overall B2B effect 
print(df.groupby("B2B")[["teamScore", "fieldGoalsPercentage", "turnovers", "win"]].mean())
```

```python
# Aggregate averages for B2B = 0 (Rest) and B2B = 1 (Back-to-Back)
overall_metrics = (df.groupby("B2B")[["teamScore", 
                                      "fieldGoalsPercentage", 
                                      "turnovers", 
                                      "win"]].mean().reset_index())
overall_metrics
```

```python
metrics = {
    "teamScore": "Points Scored",
    "fieldGoalsPercentage": "Field Goal %", 
    "turnovers": "Turnovers",
    "win": "Win Rate"
}

for var, label in metrics.items():
    plt.figure(figsize=(6,4))
    sns.barplot(
        x=["Rest","Back-to-Back"], 
        y=overall_metrics[var].values
    )
    plt.ylabel(label)
    plt.title(f"League-Wide Average {label}: Rest vs Back-to-Back")
    plt.show()

# when overall only winrate looks like its affected by back to back 
```

#### League-Wide Comparison: Rest vs Back-to-Back

We begin by comparing league-wide averages for key performance metrics under two conditions:
- B2B = 0 (games with at least one full day of rest)
- B2B = 1 (back-to-back games)

From the results:

• Points scored decreases slightly in back-to-back games.  
• Field goal percentage shows a small reduction under B2B conditions.  
• Turnovers appear relatively stable.  
• Win rate shows the clearest decline under back-to-back conditions.

Notably, while raw scoring metrics show modest differences, win probability
appears more sensitive to rest condition. This suggests that fatigue effects
may manifest more strongly in overall game outcome rather than in single
offensive efficiency metrics alone.

#### Team-Level Differences in B2B Effects

League-wide averages may mask variation across teams. Therefore,
we compute team-level average performance under both rest conditions
and calculate the difference between them.

For each team we compute:

• Points Drop = Rest average − B2B average  
• FG% Drop = Rest average − B2B average  
• Turnover Increase = B2B average − Rest average  
• Win Rate Drop = Rest average − B2B average  

Positive values indicate worse performance during back-to-back games.

```python
# comparing B2B by teams 
team_metrics = df.groupby(["teamId","B2B"])[["teamScore","fieldGoalsPercentage","turnovers","win"]].mean().reset_index()

variables = ["teamScore","fieldGoalsPercentage","turnovers","win"]

# Pivot all variables at once
team_pivot = team_metrics.pivot(index="teamId", columns="B2B", values=variables)
team_pivot.head()
```

```python
# Flatten multi-level columns (optional, easier to work with)
team_pivot.columns = [f"{var}_B2B{b2b}" for var, b2b in team_pivot.columns]
team_pivot.head()
```

```python
# Compute differences (positive values indicate performance decreased)
team_pivot["score_drop"] = team_pivot["teamScore_B2B0"] - team_pivot["teamScore_B2B1"]
team_pivot["fg_drop"] = team_pivot["fieldGoalsPercentage_B2B0"] - team_pivot["fieldGoalsPercentage_B2B1"]
team_pivot["turnover_increase"] = team_pivot["turnovers_B2B1"] - team_pivot["turnovers_B2B0"]
team_pivot["win_drop"] = team_pivot["win_B2B0"] - team_pivot["win_B2B1"]

team_pivot.head()

# one bar plot for each variable
variables = ["score_drop", "fg_drop", "turnover_increase", "win_drop"]
y_labels = ["Points Drop", "FG% Drop", "Turnovers Increase", "Win Rate Drop"]
titles = [
    "B2B Effect on Points Scored per Team",
    "B2B Effect on FG% per Team",
    "B2B Effect on Turnovers per Team",
    "B2B Effect on Win Rate per Team"
]

for var, ylabel, title in zip(variables, y_labels, titles):
    plt.figure(figsize=(12,6))
    sns.barplot(x=team_pivot.index, y=team_pivot[var])
    plt.xticks(rotation=90)
    plt.ylabel(ylabel)
    plt.title(title)
    plt.show()
```

#### Interpretation of Team-Level Results

The team-level analysis reveals that:

• Most teams experience a decline in points scored during B2B games.  
• Field goal percentage drops are generally small but consistent.  
• Turnover effects vary more across teams, suggesting strategic or stylistic differences.  
• Win rate decline is widespread and more pronounced than scoring changes.

Importantly, the magnitude of the B2B effect varies by team. Some teams
appear more resilient to scheduling congestion, while others show
substantial declines. This heterogeneity suggests that the impact of fatigue may interact
with team-specific characteristics such as depth, pace of play or

#### Multivariate Regression Analysis (OLS)

Up to this point, our analysis compared average performance between rest and back-to-back (B2B) games. While these comparisons are useful, they do not control for other important performance factors such as shooting efficiency, turnovers or home-court advantage.

To better isolate the effect of fatigue, we applied **Ordinary Least Squares (OLS) regression**. OLS estimates the relationship between a dependent variable and multiple predictors by minimizing the sum of squared residuals. In this context, OLS allows us to measure the effect of back-to-back games while controlling for other variables that influence scoring.

Formally, the model estimates:

$$
\text{teamScore} = \beta_0 
+ \beta_1(\text{B2B}) 
+ \beta_2(\text{home}) 
+ \beta_3(\text{fieldGoalsPercentage}) 
+ \beta_4(\text{turnovers}) 
+ \epsilon
$$

This approach moves our analysis from simple comparisons to a multivariate framework. Let's prepare the modeling dataset to ensure we remove first-game rows and keep only relevant predictors.

```python
# Drop first-game rows where rest_days is NaN
model_df = df.dropna(subset=["rest_days"]).copy()

# Select modeling variables
model_vars = [
    "teamScore",
    "win",
    "B2B",
    "home",
    "fieldGoalsPercentage",
    "turnovers"
]

# set model_df to the subset of columns in model_vars
model_df = model_df[model_vars]

model_df.head()
```

**Regression Model: Predicting Team Score**

Now let's see if back-to-back affect scoring AFTER controlling for performance factors by creating a regression model that predict the team score.

```python
import statsmodels.api as sm

# Define predictors
X = model_df[["B2B", "home", "fieldGoalsPercentage", "turnovers"]]
X = sm.add_constant(X)

y_score = model_df["teamScore"]

# Fit OLS model
score_model = sm.OLS(y_score, X).fit()

print(score_model.summary())
```

The regression model explains approximately 48% of the variation in team scoring (R² = 0.481). For sports data, this represents strong explanatory power.

Key Findings:

- Back-to-Back (B2B): −1.446 (p < 0.001)  
  After controlling for shooting efficiency, turnovers, and home-court advantage, teams score approximately 1.45 fewer points in back-to-back games.  
  This effect is statistically significant, indicating that fatigue has an independent impact on scoring.

- Home: +0.642 (p < 0.001)
  Teams score slightly more at home, consistent with known home-court advantage effects.

- Field Goal Percentage: +165.7 (p < 0.001)
  Because FG% is measured as a proportion (e.g., 0.45), a 1 percentage-point increase (0.01) corresponds to approximately:

  $$
  165.7 \times 0.01 \approx 1.66 \text{ additional points}
  $$

  This is practically meaningful.

- Turnovers: −0.548 (p < 0.001)
  Each additional turnover reduces scoring by roughly half a point.

Overall, even after controlling for core performance metrics, back-to-back status remains a statistically significant predictor of scoring.


**Interaction Model: Does Fatigue Differ at Home vs. Away?**

To further explore the fatigue effect, we introduced an interaction term between B2B and home. This tests whether the impact of fatigue depends on game location.

The interaction model estimates:

$$
\text{teamScore} = \beta_0 
+ \beta_1(\text{B2B}) 
+ \beta_2(\text{home}) 
+ \beta_3(\text{B2B} \times \text{home})
+ \beta_4(\text{fieldGoalsPercentage})
+ \beta_5(\text{turnovers})
+ \epsilon
$$

```python
# Create interaction term
model_df["B2B_home_interaction"] = model_df["B2B"] * model_df["home"]

X_interaction = model_df[[
    "B2B",
    "home",
    "B2B_home_interaction",
    "fieldGoalsPercentage",
    "turnovers"
]]

# intercept (bias term/columns of ones)
X_interaction = sm.add_constant(X_interaction)

# fit the model 
interaction_model = sm.OLS(model_df["teamScore"], X_interaction).fit()

print(interaction_model.summary())
```

Interpretation of the above result:

- On the road (home = 0):

  $$
  \text{Effect of B2B} = -1.885
  $$

  Back-to-back games reduce scoring by approximately 1.89 points.

- At home (home = 1):

  $$
  -1.885 + 1.144 \approx -0.74
  $$

  The fatigue effect is reduced to approximately 0.74 points.

This suggests that home-court advantage mitigates the negative impact of fatigue, but does not eliminate it entirely.


#### Interpretation and Implications

While earlier league-wide averages suggested only modest differences between rest and B2B games, the regression analysis reveals that fatigue has a statistically significant and structurally meaningful effect on scoring.

The effect size (roughly 1–2 points per game) may appear small, but in competitive NBA games where margins are often narrow, this difference can meaningfully influence outcomes.

Additionally, the interaction model provides a deeper insight: fatigue impacts teams more severely on the road. This aligns with intuitive expectations about travel strain and reduced environmental familiarity.

Overall, incorporating OLS strengthens the EDA by:

- Controlling for confounding performance variables  
- Quantifying the independent effect of fatigue  
- Revealing conditional effects (location-dependent fatigue)  
- Moving from descriptive comparison to analytical modeling  

This multivariate analysis provides stronger evidence that back-to-back scheduling has a measurable impact on team performance.
